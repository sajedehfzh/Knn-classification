  
* to evaluate the regressor, use the mean squared error instead of accuracy: 
$$ MSE(y,\hat{y}) = {1 \over n_{samples}} \sum_{i=0}^{n_{samples}-1} (y_i - \hat{y_i})^2$$

$\hat{y}$ is the predicition of the model, compared to the true value $y$.

```python
# Importing the required modules
import numpy as np
from scipy.stats import mode


# Euclidean Distance
def eucledian(p1, p2):
    dist = np.sqrt(np.sum((p1 - p2) ** 2))
    return dist


# Function to calculate KNN
def predict(x_train, y, x_input, k):
    op_labels = []

    # Loop through the Datapoints to be classified
    for item in x_input:

        # Array to store distances
        point_dist = []

        # Loop through each training Data
        for j in range(len(x_train)):
            distances = eucledian(np.array(x_train[j]), item)
            # Calculating the distance
            point_dist.append([j,distances])
#         point_dist = np.array(point_dist)

        # Sorting the array while preserving the index
        # Keeping the first K datapoints
        neighbor_arr = sorted(point_dist, key=lambda x: x[1])[:k]
        
        y_arr = [y[i[0]] for i in neighbor_arr]

        avg_y=np.mean(y_arr)
        op_labels.append(avg_y)
        

    return op_labels


# Importing the required modules
# Importing required modules
from sklearn.metrics import mean_squared_error
from numpy.random import randint

data_set = np.load('dataset_4.npz')
X=data_set['X']
y=data_set['y']


# Creating the training Data
train_idx = xxx = randint(0, 100, 75)
X_train = X[train_idx]
y_train = y[train_idx]

# Creating the testing Data
test_idx = xxx = randint(0, 100, 25)  # taking 25 random samples
X_test = X[test_idx]
y_test = y[test_idx]

# Applying our function
y_pred = predict(X_train, y_train, X_test, 5)
print("Mean squared error : ",mean_squared_error(y_test,y_pred)*100)
plt.scatter(X,y)
plt.scatter(X_test,y_pred,c="red")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
